```python


###import modules

import pandas as pd
from tkinter import *
from matplotlib.backends.backend_tkagg import (
    FigureCanvasTkAgg, NavigationToolbar2Tk)

### Implement the default Matplotlib key bindings.
from matplotlib.backend_bases import key_press_handler
from matplotlib.figure import Figure
import matplotlib.pyplot as plt

import numpy as np

###Import star data from files scraped from wikipedia from two files

df = pd.read_csv('data_stars_TW.csv', sep=';', na_values= ".")
df_m = pd.read_csv('data.csv')
df['alter'] = df['death'] - df['birth']
###Drop missing data 
df.dropna(inplace = True)
df_m.dropna(inplace =True)

###Cleaning of citation brackets from datafile
def clean_klma(x):
    m = ""
    flag = True
    for i in x:
        if i == "[":
            flag = False
        elif i == "]":
            flag = True
            continue
        if flag:
            m += i
    return m

df_m['data'] = df_m.data.apply(clean_klma)


###Function executed when Button "Show" is used
def button_action():
    ###Fetch values from the input fields
    year = e4.get()
    month = variable_m.get()
    day = e6.get()
    month_day = month + " " +day
    wahl = variable.get()
    ### User can input for day two different formats - 01 and 1 for the first of a given month
    if day in ['01','02','03','04','05','06','07','08','09']:
        day = day[1]

    ### Combining separate entries into a full combination of month//day
    month_day = month + " " +day
    message= "Ups, no matches!"  
    m = ""
    ###to check if data is entered into the entry fields for birthday - otherwise error message
    if wahl == u1:
        df_sub = df_m[df_m['date'] == month_day]
        zahl = len(df_sub)
        if zahl != 0:
            message = zahl
    elif wahl == u2:
        df_sub = df_m[(df_m['date'] == month_day)&(df_m['year']==year)]
        zahl = len(df_sub)
        if zahl != 0:
            for element  in df_sub['data'].values:
                m = element
        message = m
    elif wahl == u3:
        df_sub = df_m[df_m['date'] == month_day]
        df_jung = df_sub.tail()
        zahl = len(df_sub)
        if zahl != 0:
            for element  in df_jung['data'].values:
                m += element + "\n"
            message = m
    elif wahl == u4:
        df_sub = df_m[df_m['date'] == month_day]
        df_alte = df_sub.head()
        zahl = len(df_sub)
        if zahl != 0:
            for element  in df_alte['data'].values:
                m += element + "\n"
            message = m
    ###Show message command           
    ausgabe.configure(text=message)

###Function executed when Button "quit" is used
def close(): 
    fenster.destroy() 

###Function executed when Button "Year of Birth vs. Age" is used ot create a new window
def createNewWindow():
    window = Toplevel(fenster)
    
    ###Match dropdown string value for month with numerical value for datatset call
    month=  variable_m.get()
    day=  e6.get()  
    if month == "January":
        month =1.0
    elif month == "Februar":
        month =2.0
    elif month == "March":
        month =3.0
    elif month == "April":
        month =4.0
    elif month == "May":
        month =5.0
    elif month == "June":
        month =6.0
    elif month == "July":
        month =7.0
    elif month == "August":
        month =8.0
    elif month == "September":
        month =9.0
    elif month == "October":
        month =10.0
    elif month == "November":
        month =11.0
    elif month == "December":
        month =12.0
    if day in ['01','02','03','04','05','06','07','08','09']:
        day = day[1]
    day = float(day)

    ###Create subdatasets of stars_data where column month and column day match the input value by the user
    data=df[(df['month'] == month) & (df['day'] == day)]
    
    ###Create plot for Birthyear vs. Age
    fig = plt.figure(figsize=(4, 5))
    plt.scatter(x="birth",y="alter", alpha=0.5, data=data)
    plt.title('Year of Birth vs. Age')
    plt.xlabel('Year of Birth')
    plt.ylabel('Age')
    ### specify the window as master
    canvas = FigureCanvasTkAgg(fig, master=window)
    canvas.draw()
    canvas.get_tk_widget().grid(row=1, column=0, ipadx=40, ipady=20)
###Create an object as window with title 
fenster = Tk()
fenster.title("Promi Finder")

####Label for the data entry fields
Label(fenster, text="First name", font=("Helvetica", 14)).grid(row=1)
Label(fenster, text="Middle name", font=("Helvetica", 14)).grid(row=2)
Label(fenster, text="Last name", font=("Helvetica", 14)).grid(row=3)
Label(fenster, text="Year of birth", font=("Helvetica", 14)).grid(row=4)
Label(fenster, text="Month of birthday", font=("Helvetica", 14)).grid(row=5)
Label(fenster, text="Birthday", font=("Helvetica", 14)).grid(row=6)

Label(fenster, justify = CENTER, font=("Helvetica", 14),
             text="""\n
************** Find your star(s) **************""").grid(row=11, column = 0, columnspan = 3)

###Selection for analysis between ones own birthday and the stars birthday database
u1 = "Stars with the same birthday"
u2 = "Stars with the exact same birthday"
u3 = "5 younger stars with the same birthday"
u4 = "5 older stars with the same birthday"
###Output in Promi Finder
variable = StringVar(fenster)
variable.set(u1) # default value

optionen = OptionMenu(fenster, variable, u1, u2, u3, u4)
optionen.configure(width = 40, font=("Helvetica", 14))
###Variable assignement to the data entry windows
e1 = Entry(fenster)
e2 = Entry(fenster)
e3 = Entry(fenster)
e4 = Entry(fenster)

m1 = "January"
m2 = "Februar"
m3 = "March"
m4 = "April"
m5 = "May"
m6 = "June"
m7 = "July"
m8 = "August"
m9 = "September"
m10 = "October"
m11 = "November"
m12 = "December"

variable_m = StringVar(fenster)
variable_m.set(m1) # default value

optionen_m = OptionMenu(fenster, variable_m, m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11,m12)
optionen_m.configure(width = 20, font=("Helvetica", 14))

e6 = Entry(fenster)

e1.grid(row=1, column=1)
e2.grid(row=2, column=1)
e3.grid(row=3, column=1)
e4.grid(row=4, column=1)
optionen_m.grid(row=5, column=1)
e6.grid(row=6, column=1)

###Define the buttons and linkage with functions
optionen.grid(row = 12, column = 0, columnspan = 4, pady = 20)
Button(fenster, text="Quit", font=("Helvetica", 14), command=close).grid(row=13, column=0, pady=4)
photo=PhotoImage(file='Walk-Fame-Screenshot.png')
label1 = Label(fenster, justify = CENTER, font=("Helvetica", 14), image = photo)
new_window = Button(fenster,text="Year of Birth vs. Age", font=("Helvetica", 14), command = createNewWindow)
ausgabe = Label(fenster, justify = LEFT, font=("Helvetica", 14), text=" ")
run_button = Button(fenster,text="Show", font=("Helvetica", 14), command = button_action)
run_button.grid(row = 13, column = 2)
label1.grid(row = 0, columnspan=3)
ausgabe.grid(row = 15, column = 0, columnspan = 3)
new_window.grid(row=13, column=1)

###Execute main program for the GUI
mainloop()

```
