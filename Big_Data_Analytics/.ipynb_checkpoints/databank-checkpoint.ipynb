```python
def create_constr(db_schema, db_typ='mysql+mysqlconnector', db_user='root', db_password='', db_adr='localhost', echo=0):
    '''Diese Funktion liefert uns einen Connection-String zurück.'''
    try:
        str_con = f'{db_typ}://{db_user}{db_password}@{db_adr}/{db_schema}'
        if echo == 1:
            print(f'Dieser Connection-String wurde erzeugt: {str_con}\n')
        return str_con 
    except:    
        #raise
        print('Etwas hat nicht funktioniert.')
        return ''
```

```python
str_sql = "select *\
    from penn pe \
    left join polity po \
    on (po.country = pe.country ) and (po.year = pe.year) "
pd.read_sql(str_sql, engine).info()
```

```python
#calling some columns with joining two tables, and  then making scatter plot
str_sql = "select pe.country, pe.year, hc, polity2, rgdpo /pop as per_capita \
    from penn pe \
    join polity po \
    on (po.country = pe.country ) and (po.year = pe.year) "
pd.read_sql(str_sql, engine).plot.scatter(x='per_capita',y='polity2')
```

```python
#calling some columns with joining two tables, and  then making scatter plot
str_sql = "select polity2, rgdpo /pop as per_capita \
    from penn pe \
    left join polity po \
    on (po.country = pe.country ) and (po.year = pe.year) \
    where pe.year = 2010"
sns.lmplot(data = pd.read_sql(str_sql, engine), y= 'polity2', x='per_capita')
plt.xscale('log')
```

```python
#calling some columns with joining two tables, and  then making scatter plot
str_sql = "select pe.hc, po.polity2  \
    from penn pe \
    left join polity po \
    on (po.country = pe.country ) and (po.year = pe.year) "
pd.read_sql(str_sql, engine).plot.scatter(x='hc',y = 'polity2')
```

```python
#calling some columns with joining two tables, and then using case statement to make scatter plot
str_sql = "select pe.hc, s.Status,\
    case \
        when s.Status = 'F' then 3\
        when s.Status = 'NF' then 1\
        when s.Status = 'PF' then 2\
        else Null\
    end as pol_r\
    from penn pe \
    left join freedom_house s \
    on (s.country = pe.country ) and (s.year = pe.year) "
pd.read_sql(str_sql, engine).plot.scatter(x='hc',y='pol_r')
```

```python
str_sql = "select s.Status,\
    case \
        when s.Status = 'F' then 3\
        when s.Status = 'NF' then 1\
        when s.Status = 'PF' then 2\
        else Null\
    end as pol_r\
    from penn pe \
    left join freedom_house s \
    on (s.country = pe.country ) and (s.year = pe.year) "
sns.countplot(pd.read_sql(str_sql, engine)['pol_r'])
```

```python
#groupping countries to get average value
str_sql = "select country, avg(rgdpo) as avg_gdp\
    from penn\
    group by country\
    order by avg_gdp"
pd.read_sql(str_sql, engine)
```

```python
#groupping by status after joining two tables
str_sql = "select pe.hc, w.life_expectancy, s.Status \
    from penn pe \
    join world_bank w\
    on (w.countrycode = pe.countrycode ) and (w.year = pe.year)\
    join freedom_house s\
    on (s.country = w.country ) and (s.year = w.year) "
pd.read_sql(str_sql, engine).groupby('Status').mean()
```

```python
#using where clauses to get line plot for each status
for i in ['F','NF','PF']:
    str_sql =f"select pe.hc, w.life_expectancy, s.Status \
    from penn pe \
    join world_bank w\
    on (w.countrycode = pe.countrycode ) and (w.year = pe.year)\
    join freedom_house s\
    on (s.country = w.country ) and (s.year = w.year)\
    where s.Status = '{i}'"
    sns.lmplot(data = pd.read_sql(str_sql, engine), x='hc', y='life_expectancy')
    plt.title(i)
    plt.show()
```
