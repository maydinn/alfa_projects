```python
def drop_multi_corr(df, target, value=0.7):
    df_c = df.copy()
    df_c = df_c._get_numeric_data()
    df_t =df_c[[target]]
    df_c.drop(target, 1, inplace=True)
    drop_list=[]
    for i in range(len(df_c.columns)-2):
        a = i
        for j in range(i,len(df_c.columns)-2):
             
            num = df_c.iloc[:,i].corr(df_c.iloc[:,j+1])

            if (num > value) or (num < -value):
                if df_t[target].corr(df_c.iloc[:,i])*df_t[target].corr(df_c.iloc[:,i]) < df_t[target].corr(df_c.iloc[:,j+1])*df_t[target].corr(df_c.iloc[:,j+1]):
                    drop_list.append(df_c.columns[i])
                    print(df_c.columns[i])

                else:
                    drop_list.append(df_c.columns[j+1])
                    print(df_c.columns[j+1])

    df_c.drop(columns= drop_list, inplace=True)
    
    return pd.concat([df_c,df_t,df.drop(columns = df._get_numeric_data().columns)],1)
```

```python
def classification_f1(df, target, decisiontree = True, randromforest=True, logit=True, kmean=(True,40), xgb=True, svc=True):
    df_c =df.copy()
    X_train_s, X_test_s, y_train, y_test = train_test_split( df_c.drop(target,1),df_c[target], test_size=0.20 ,random_state=42)

    
    if decisiontree:
        tree = DecisionTreeClassifier()
        tree.fit(X_train_s, y_train)
        y_pred = tree.predict(X_test_s)
        tree_f = f1_score(y_test,y_pred, average='weighted')
    if randromforest:
        rand = RandomForestClassifier()
        rand.fit(X_train_s, y_train)
        y_pred = rand.predict(X_test_s)
        rand_f = f1_score(y_test,y_pred, average='weighted')
    if logit:
        log = LogisticRegression()
        log.fit(X_train_s, y_train)
        y_pred = log.predict(X_test_s)
        log_f = f1_score(y_test,y_pred, average='weighted')
    if xgb:
        xf = XGBClassifier()
        xf.fit(X_train_s, y_train)
        y_pred = xf.predict(X_test_s)
        xf_f = f1_score(y_test,y_pred, average='weighted')
    if kmean[0]:
        acc = []
        for i in range(1, kmean[1]):
            neigh = KNeighborsClassifier(n_neighbors = i).fit(X_train_s,y_train)
            yhat = neigh.predict(X_test_s)
            acc.append(accuracy_score(y_test, yhat))
        n = acc.index(max(acc)) + 1
        neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train_s,y_train)
        y_pred = neigh.predict(X_test_s)  
        km_f = f1_score(y_test,y_pred, average='weighted')
    if svc:
        sv = SVC()
        sv.fit(X_train_s, y_train)
        y_pred = sv.predict(X_test_s)
        sv_f = f1_score(y_test,y_pred, average='weighted')

    return sns.barplot(x=[tree_f, rand_f, log_f, xf_f, km_f, sv_f], \
                       y=['Decision Tree', 'RandomForest','Logistic Reg','XGB Bost','KNeighborsClassifier','SVM'])
```

```python
import numpy.linalg as la 
from scipy import stats 
import numpy as np
```

```python
def hausman(fe, re): 
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    a = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b - B).T, la.inv(v_b - v_B).dot(b - B))
    pval = stats.chi2.sf(chi2, a)
    return chi2, a, pval
hausman_results = hausman(fe, re) 
print('chi-Squared: ' + str(hausman_results[0]))
print('degrees of freedom: ' + str(hausman_results[1])) 
print('p-Value: ' + str(hausman_results[2]))
```
